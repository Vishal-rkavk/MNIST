```python
import tensorflow as tf 
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
```

```python
x_train.shape
```

```python
#Turning the 2D array to 1D array, to ffed to neural network
x_train_flat = x_train.reshape(len(x_train),28*28)
x_test_flat  = x_test.reshape(len(x_test),28*28)
```

```python
#Scaling the features
x_train_flat = x_train_flat/255
x_test_flat = x_test_flat/255
```

```python
model = keras.Sequential([
     keras.layers.Dense(10,input_shape = (784,),activation="sigmoid")
 ])
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy']
             )
model.fit(x_train_flat,y_train,epochs=5)
```

```python
model.evaluate(x_test_flat,y_test)
```

```python
y_predict = model.predict(x_test_flat)
```

```python
#y_predictions = [np.argmax(i) for i in y_predict]
y_prediction = []
```

```python
for x in y_predict:
    y_prediction.append(np.argmax(x)) 
```

```python
#confusion matrix
cm = tf.math.confusion_matrix(labels= y_test, predictions=y_prediction)
```

```python
import seaborn as sns
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(cm,annot=True,ax=ax,fmt='d')
plt.xlabel("predicted")
plt.ylabel("truth")
```

```python
#Adding hidden layer to make model better

model = keras.Sequential([
     keras.layers.Dense(100,input_shape = (784,),activation="relu"),
     keras.layers.Dense(10,activation="sigmoid")
    
 ])
model.compile(optimizer='SGD',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy']
             )
model.fit(x_train_flat,y_train,epochs=5)
```

```python
model.evaluate(x_test_flat,y_test)
```
